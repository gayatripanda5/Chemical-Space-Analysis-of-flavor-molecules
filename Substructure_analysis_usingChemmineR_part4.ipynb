```r
### Substructure Analysis for CGS mini project:
library("ChemmineR")
library("fmcsR")
library("ape")

#setwd("~/Courses/CGS_Project")
```

```r
# loading the sdf file - generated using openbabel gui
AI <- read.SDFset("Actives_Inactives.sdf")
```

```r
# calculating the pairwise similarity matrix for all the molecules
d1 <- data.frame(sapply(cid(AI), function(x) fmcsBatch(AI[x], AI, au=0, bu=0, matching.mode="aromatic")[,"Overlap_Coefficient"]))

write.csv(d1, file = "Similarity_Matrix_Actives_Inactives.csv")
```

```r
head(d1)
```

```r
# Performing heirarchial clustering on the above generated matrix
hc1 <- hclust(as.dist(1-d1), method = "complete")
```

```r
head(hc1)
```

```r
# Plotting dendrogram after heirarchial clustering 
plot(as.dendrogram(hc1), edgePar = list(col = 4, lwd = 2, horiz = TRUE))
```

```r
# We want circular dendrogram with the names of the molecules as well as the colors for their respective groups i.e. Active or Inactive. Therefore, we save the following tree as .newick file which will then be used to create a dendrogram using iTOL, an open source online tool to render phylogeny
my_tree1 <- as.phylo(hc1) 
write.tree(phy = my_tree1, file="Tree_Actives_Inactives.newick")
```
